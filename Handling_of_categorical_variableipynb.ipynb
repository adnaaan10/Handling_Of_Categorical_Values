```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/customer.csv')
df
```

# **ORDINAL ENCODING FOR INDIPENDENT VARIABLE**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
review_order=['Poor','Average','Good']
education_order=['School','UG','PG']
encoder=OrdinalEncoder(categories=[review_order,education_order])
```

```python
df[['review','education']]=encoder.fit_transform(df[['review','education']])
```

```python
df
```

# **LABEL ENCODING FOR TARGET VARIABLE**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['purchased']=LabelEncoder().fit_transform(df['purchased'])
```

```python
df
```
